```python
 import matplotlib.pyplot as plt
 import numpy as np
 import pandas as pd
```

```python
x = np.array([2, 4])
y = np.array([5, 15])

plt.plot(x, y)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.figure(figsize=(8,5))
plt.plot(x,y,color='blue',linestyle=':',marker='s',linewidth=3)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("line chart")
plt.legend(['my data series'])
plt.grid(True)
plt.show()
```

```python
categories=['A','B','C','D','E']
values=[5,2,3,4,6]
plt.figure(figsize=(8,5))
plt.bar(categories,values,color='green')
plt.xlabel('categories')
plt.ylabel('values')
plt.title('Bar chart')
plt.show()
```

```python

data = [12, 15, 13, 17, 16, 16, 19, 12, 14, 16, 14, 17, 18, 18, 15, 14]


plt.hist(data, bins=16, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Simple Histogram Example')
plt.show()

```

```python
x = [1, 2, 3, 4, 5]
y = [5, 7, 4, 6, 8]


plt.scatter(x, y)
plt.xlabel('X Value')
plt.ylabel('Y Value')
plt.title('Simple Scatter Plot Example')
plt.show()
```

```python

labels = ['cherry', 'Bindu', 'rupa', 'D']
sizes = [15, 30, 45, 10]


plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Simple Pie Chart Example')
plt.axis('equal')
plt.show()
```

```python
data = [7, 8, 5, 6, 7, 9, 12, 5, 7, 8, 6]


plt.boxplot(data)
plt.title('Simple Box Plot Example')
plt.ylabel('Values')
plt.show()
```

```python

data = np.random.rand(10, 10)


plt.imshow(data, cmap='viridis', interpolation='nearest')
plt.colorbar()
plt.title('Simple Heatmap Example')
plt.show()
```
