```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=np.array([3,9])
y=np.array([4,7])
plt.plot(x,y)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.figure(figsize=(8,5))
plt.plot(x,y,color='black',linestyle='-.',marker='d',linewidth=1)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("line chart")
plt.legend(['my data series'])
plt.grid(True)
plt.show()
```

```python
categories=['summer','rainy','winter','spring','autumn']
values=[9,3,5,10,7]
plt.figure(figsize=(8,5))
plt.bar(categories,values,color='lavender')
plt.xlabel('seasons')
plt.ylabel('rating')
plt.title('Seasons')
plt.show()
```

```python
data = [12, 15, 13, 17, 16, 16, 19, 12, 14, 16, 14, 17, 18, 18, 15, 14]


plt.hist(data, bins=16,edgecolor='black', color='lavender')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title(' Histogram ')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [6, 7, 8, 9, 10]
plt.scatter(x, y,color='red',marker='*')
plt.xlabel('X Value')
plt.ylabel('Y Value')
plt.title('Simple Scatter Plot Example')
plt.show()
```

```python
labels = ['ice cream', 'chocolate', 'cookies', 'candies','cupcakes','mochi','apple pie']
sizes = [50, 30, 45, 10,15,10,20]
colors=['lavender','pink','lightgreen','lightyellow','grey','lightblue']
plt.pie(sizes,colors=colors, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Desserts')
plt.axis('equal')
plt.show()
```

```python

data = [7, 8, 5, 6, 7, 9, 12, 5, 7, 8, 6]


plt.boxplot(data)
plt.title('Syringe')
plt.ylabel('')
plt.show()
```

```python
data = np.random.rand(10, 10)
plt.imshow(data, cmap='ocean', interpolation='nearest')
plt.colorbar()
plt.title('Simple Heatmap Example')
plt.show()

```
